## Manipulating Strings 
I do not like how the two columns, genre and production_companies look. I will look at their data type first. Once I have that information I will know how to work with them. Lets look at genres first:

## Changing Genres Data Type

```python
# This function is used in python to determine the type of the data you have in the ()
print(type(df['genres'][1]))
print(df['genres'][1])
print(df['genres'][1][3])
```

```python
# It looks like a dictionary is being held as a string (because its in the "") 

# This is why when I sliced into the string it returns the letter 'i'

# I want it to be a dictionary so I can pull the 'name' out and set that as the genre vs a string of dictionary

ast.literal_eval(df['genres'][1]) # This is a test

# Within the line of code above we used the ast and literal_eval

# ast 'process trees of the Python abstract syntax grammar' (refer to link in resources list)

# literal_eval in ths case is eveluating the str to then returns what inside of it which is a dictionary
```

```python
# How do I get this applied to every index in the df of the genres column?
# Function maybe

'''This function will take in a string as an argument and evaluate it.
The outcome of the evaluation is assigned to a variable conv_list.
The variable represents a list of Dictionaries.
The funtion then uses a 1 time use lambda function to filter out everything but 
the value of the 'name' key in the dictionary.
That is now the new conv_list and is the result of the function.'''  

def convert_it(string):  
        conv_list = ast.literal_eval(string)
        conv_list = list(map(lambda x: x['name'],conv_list))
        return conv_list
```

```python
# Checking to see if the function works 

convert_it(df['genres'][0])
```

```python
# Assigning the outcome of the applied function to my genres column
df.genres = df.genres.map(lambda x: convert_it(x))
```

```python
# Looking to see how the column is formated after the function
df.genres.head()
```

## Changing Production_Companies Data Type


```python
# Now apply the same thing done to genres but to production_companies by using the same function
# we used for genres

# Assigning the outcome of the applied function to my production_companies column
df.production_companies = df.production_companies.map(lambda x: convert_it(x))
```

```python
# Looking to see how the column is formated after the function
df.production_companies.head()
```
