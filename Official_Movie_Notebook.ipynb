## Time to Clean Up

```python
# Time to find the 13 rows that are being repeated 
df.loc[df.duplicated()]
```

```python
# Now drop the duplicates but keep the repeats
# But keep the first occurrence of the row
df.drop_duplicates(keep='first', inplace=True) # remeber to put inplace true so apply and SAVE changes to DataFrame
```

```python
# Check to see if you have any repeats
df.loc[df.duplicated()]
```

## Dropping Columns 


```python
# Dropping the homepage column 
df.drop(columns= 'homepage', inplace=True)
```

```python
# Checking my work
df.columns

# What do you know!! Homepage column is no longer in the dataframe
```

```python
# Here I grabbed all of the columns I didn't want and put them a list 
del_columns = ['status', 'poster_path', 'overview', 'original_title', 'original_language', 'imdb_id', 'belongs_to_collection', 'tagline']
```

```python
del_columns
```

```python
# Bring list of columns I made above down to this cell to be used in my drop function
df.drop(columns = del_columns, inplace=True)
```

```python
# Let's check to see if it worked
df.columns

# When this ran all the columns in the del_columns list were removed for the df perminantly 
```
