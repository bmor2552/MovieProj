## Executing the Plan

## An Issue Has Been Brought to My Attention


The merge worked! But Something isn't right....
The data I thought I was going to have had over 10,000 rows BUT they movies in the rows were being repeated!! 
So I have to drop the movies that keep repeating in my DataFrame. At the end of this I should have the true size and shape of my newly merged DataFrame.

Here goes nothing:

```python
# When I run this I see more than 1 row that has this movie id! 
# I do not want that because it will mess up my analysis of the data

df_merge.loc[df_merge['id'] == 546554].shape
```

## Drop Repeats

```python
# The repeats will be dropped here  
# I will assign this to a variable
# This is the first step I have taken to clean my data. 
df_new_merged = df_merge.drop_duplicates(keep= 'first')
```

## Conclusion 

Unfortunatly this is not enough data for me to use for my project. 
Why did my data go from over 7,000 rows to 460? 
Well, when pulling my data I failed to realize that I was grabbing data based off of year for my time frame vs the release date

Ex: Batman comes out in 2010 right, but a group of people have voted this as their favorite movie for the past 10 years. My API criteria doesn't know that this is happening and just pulls from 2010 to 2020 the movies that were voted the most popular for the every year during a 10yr time frame. That will mean if enough people voted for Batman as their favorite movie in the time frame I chose, the movie will repeat itself for more than one year. Now imagine this happening for more than one movie over a 10yr span.

This is why I had so much data in the beginning amd so little data at the end.
